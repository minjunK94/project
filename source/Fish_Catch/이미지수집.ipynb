```python
import os
import time
import json
from requests import get, exceptions

def save_img_split(idx, url, base_folder, keyword):
    """
    인덱스에 따라 train 혹은 val 폴더에 이미지를 다운로드 후 저장한다
    """
    # 인덱스 기준 폴더 설정
    if idx <= 50:
        subfolder = "train"
    else:
        subfolder = "val"

    # 최종 폴더 경로 설정
    folder_path = os.path.join(base_folder, subfolder, keyword)
    os.makedirs(folder_path, exist_ok=True)

    file_name = f"{keyword}_{idx}.jpg"
    file_path = os.path.join(folder_path, file_name)

    try:
        response = get(url, verify=False, allow_redirects=True)
        response.raise_for_status()
        if response.status_code == 200:
            with open(file_path, "wb") as f:
                f.write(response.content)
            print(f"{file_path} 저장 완료!")
        else:
            print(f"이미지 다운로드 실패: {response.status_code} - {url}")
    except exceptions.TooManyRedirects:
        print(f"TooManyRedirects 에러 발생: {url}")
    except exceptions.RequestException as e:
        print(f"이미지 다운로드 중 에러 발생: {e}")

def download_images(keywords, client_id, client_secret):
    base_folder = "img"
    os.makedirs(base_folder, exist_ok=True)

    for keyword in keywords:
        print(f"'{keyword}' 키워드 이미지 다운로드 중...")

        # 네이버 API 호출
        url = f'https://openapi.naver.com/v1/search/image?query={keyword}&display=60'
        headers = {
            "X-Naver-Client-Id": client_id,
            "X-Naver-Client-Secret": client_secret
        }

        time.sleep(3)

        try:
            response = get(url, headers=headers, allow_redirects=True)
            response.raise_for_status()
            items = json.loads(response.text).get('items', [])

            for idx, item in enumerate(items[:61]):
                image_url = item['link']
                time.sleep(1)
                # save_img(f"{keyword}_{idx}", image_url, keyword_folder)
                save_img_split(idx, image_url, base_folder, keyword)

            print(f"'{keyword}' 키워드 이미지 다운로드 완료!")
        except exceptions.TooManyRedirects:
            print(f"TooManyRedirects 에러 발생: {url}")
        except exceptions.RequestException as e:
            print(f"API 요청 중 에러 발생: {e}")

    print("모든 키워드에 대한 이미지 다운로드 완료!")
# 키워드 리스트
keywords = ['강도다리', '문치가자미', '돌가자미', '감성돔', '뱀장어', '미유기', '뚝지', '전복',
            '대농갱이', '메기', '동자개', '은어', '쏘가리', '붕어', '넙치', '쥐노래미', '대구', '명태',
            '대문어', '참게', '연어', '해삼', '갑오징어', '미꾸라지', '미꾸리', '대하', '빙어', '꽃게',
            '점농어', '황복', '조피볼락', '주꾸미', '동남참게', '버들치', '보리새우', '볼락', '잉어',
            '꼼치', '돌돔', '전어', '왕우럭', '말쥐치', '참돔', '물렁가시붉은새우', '개볼락',
            '홍해삼','흑해삼','톱날꽃게','쥐치','황어','참조기','부세','낙지','능성어','민어',
            '박대','자바리','쏨뱅이']

# 함수 실행
download_images(keywords, client_id, client_secret)
```
