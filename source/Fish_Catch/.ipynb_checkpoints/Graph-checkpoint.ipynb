```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# warning(경고) 안보이게
import warnings
warnings.filterwarnings(action='ignore') # 경고메세지 안보이게

# 시각화의 선명도를 높임
%config InlineBackend.figure_format = 'retina'

# 한글설정
plt.rc('font', family='Malgun Gothic') # 윈도우즈
plt.rc('axes', unicode_minus=False) # 축의 - 깨짐 방지
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('data/어획_방류.csv')

# Filter data for the year 2023
data_2023 = data[data['연도'] == 2023]

# Aggregate data by 시도 and 품종
summary_2023 = data_2023.groupby(['시도', '품종']).agg({
    '어획량_마리': 'sum',
    '어획량_Kg': 'sum',
    '방류량': 'sum'
}).reset_index()

unique_cities = summary_2023['시도'].unique()

for city in unique_cities:
    city_data = summary_2023[summary_2023['시도'] == city]

    # Create a figure for each 시도
    plt.figure(figsize=(16, 8))
    
    # Plot 어획량 (마리)
    bar1 = sns.barplot(data=city_data, x='품종', y='어획량_마리', color='skyblue', label='어획량 (마리)')
    
    # Plot 방류량
    bar2 = sns.barplot(data=city_data, x='품종', y='방류량', color='lightgreen', alpha=0.7, label='방류량')

    # Add title and labels
    plt.title(f'{city} 시도 - 2023년도 품종별 어획량 및 방류량', fontsize=16)
    plt.xlabel('품종', fontsize=14)
    plt.ylabel('수량', fontsize=14)
    plt.xticks(rotation=45)
    plt.legend(fontsize=12, loc='upper right')
    
    plt.tight_layout()
    plt.show()
```

```python
# Filter data for 2023 and 2024 separately
data_2023 = data[data['연도'] == 2023]
data_2024 = data[data['연도'] == 2024]

# Aggregate data by 시도 and 품종 for each year
summary_2023 = data_2023.groupby(['시도', '품종']).agg({
    '어획량_마리': 'sum',
    '방류량': 'sum'
}).reset_index()
summary_2023['연도'] = 2023

summary_2024 = data_2024.groupby(['시도', '품종']).agg({
    '어획량_마리': 'sum'
}).reset_index()
summary_2024['연도'] = 2024

# Combine the data for 2023 and 2024 어획량
combined_catch_data = pd.concat([
    summary_2023[['시도', '품종', '어획량_마리', '연도']],
    summary_2024[['시도', '품종', '어획량_마리', '연도']]
])

# Add 방류량 information only for 2023
combined_catch_data = combined_catch_data.merge(
    summary_2023[['시도', '품종', '방류량']],
    on=['시도', '품종'],
    how='left'
)

# Create a FacetGrid for visualization
g = sns.FacetGrid(combined_catch_data, col="시도", col_wrap=4, height=5, sharey=False)

# Map barplots for 어획량_마리 and 방류량
g.map_dataframe(
    sns.barplot, 
    x="품종", 
    y="어획량_마리", 
    hue="연도", 
    palette="coolwarm", 
    dodge=True
)
g.map_dataframe(
    sns.barplot, 
    x="품종", 
    y="방류량", 
    color="lightgreen", 
    alpha=0.5, 
    label="방류량 (2023년)"
)

# Adjust plot aesthetics
g.set_titles("{col_name} 시도")
g.set_axis_labels("품종", "수량")
g.set_xticklabels(rotation=45)
g.add_legend(title="연도 / 방류량")

# Show the plot
plt.tight_layout()
plt.show()
```
