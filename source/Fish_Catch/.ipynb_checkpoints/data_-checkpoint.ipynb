```python
import pandas as pd
import numpy as np
```

```python
# Data read
# 어획량 Data
fish_catch = pd.read_csv('data/어획량_수정.csv')
# 방류 Data
fish_discharge = pd.read_csv('data/방류실적_수정.csv')
```

```python
fish_discharge
```

```python
fish_catch = fish_catch.set_index('어종류')
fish_discharge = fish_discharge.set_index('품종')
```

```python
final_fish = pd.concat([fish_catch, fish_discharge], 
                       axis=0, 
                       ignore_index=True)
```

```python
final_fish = final_fish.set_index('어종류')
final_fish
```
