```python
import os
import time
import json
from requests import get, exceptions

# 한글 키워드와 영어 번역 매핑
keyword_translations = {
    '강도다리': 'Flatfish', '문치가자미': 'Flounder', '돌가자미': 'Sole', '감성돔': 'Black Sea Bream',
    '뱀장어': 'Eel', '미유기': 'Mackerel', '뚝지': 'Gudgeon', '전복': 'Abalone',
    '대농갱이': 'Giant Sea Bass', '메기': 'Catfish', '동자개': 'Loach', '은어': 'Sweetfish',
    '쏘가리': 'Bass', '붕어': 'Carp', '넙치': 'Turbot', '쥐노래미': 'Rockfish',
    '대구': 'Cod', '명태': 'Pollock', '대문어': 'Giant Octopus', '참게': 'Freshwater Crab',
    '연어': 'Salmon', '해삼': 'Sea Cucumber', '갑오징어': 'Cuttlefish', '미꾸라지': 'Mudfish',
    '미꾸리': 'Loach', '대하': 'Prawn', '빙어': 'Icefish', '꽃게': 'Crab',
    '점농어': 'Spotted Sea Bass', '황복': 'Yellow Puffer', '조피볼락': 'Rock Bream',
    '주꾸미': 'Webfoot Octopus', '동남참게': 'Northeast Freshwater Crab', '버들치': 'Bitterling',
    '보리새우': 'Shrimp', '볼락': 'Rockfish', '잉어': 'Carp', '꼼치': 'Gudgeon',
    '돌돔': 'Dab', '전어': 'Mackerel', '왕우럭': 'Black Rockfish', '말쥐치': 'Giant Sea Bream',
    '참돔': 'Soft-shelled Turtle', '물렁가시붉은새우': 'Red Shrimp', '개볼락': 'Gudgeon',
    '홍해삼': 'Red Sea Cucumber', '흑해삼': 'Black Sea Cucumber', '톱날꽃게': 'Flower Crab',
    '쥐치': 'Gudgeon', '황어': 'Yellow Croaker', '참조기': 'Sea Bream', '부세': 'Bream',
    '낙지': 'Octopus', '능성어': 'Giant Sea Bass', '민어': 'Giant Sea Bream',
    '박대': 'Flatfish', '자바리': 'Grouper', '쏨뱅이': 'Scorpionfish'
}

def save_img_split(idx, url, base_folder, keyword, english_keyword):
    if idx <= 50:
        subfolder = "train"
    else:
        subfolder = "val"

    folder_path = os.path.join(base_folder, subfolder, english_keyword)
    os.makedirs(folder_path, exist_ok=True)

    file_name = f"{english_keyword}_{idx}.jpg"
    file_path = os.path.join(folder_path, file_name)

    try:
        response = get(url, verify=False, allow_redirects=True)
        response.raise_for_status()
        if response.status_code == 200:
            with open(file_path, "wb") as f:
                f.write(response.content)
            print(f"{file_path} 저장 완료!")
        else:
            print(f"이미지 다운로드 실패: {response.status_code} - {url}")
    except exceptions.TooManyRedirects:
        print(f"TooManyRedirects 에러 발생: {url}")
    except exceptions.RequestException as e:
        print(f"이미지 다운로드 중 에러 발생: {e}")

def download_images(keywords, client_id, client_secret):
    base_folder = "img"
    os.makedirs(base_folder, exist_ok=True)

    for keyword in keywords:
        english_keyword = keyword_translations[keyword]
        print(f"'{keyword}' ({english_keyword}) 키워드 이미지 다운로드 중...")

        url = f'https://openapi.naver.com/v1/search/image?query={keyword}&display=60'
        headers = {
            "X-Naver-Client-Id": client_id,
            "X-Naver-Client-Secret": client_secret
        }

        time.sleep(3)

        try:
            response = get(url, headers=headers, allow_redirects=True)
            response.raise_for_status()
            items = json.loads(response.text).get('items', [])

            for idx, item in enumerate(items[:61]):
                image_url = item['link']
                time.sleep(1)
                save_img_split(idx, image_url, base_folder, keyword, english_keyword)

            print(f"'{keyword}' ({english_keyword}) 키워드 이미지 다운로드 완료!")
        except exceptions.TooManyRedirects:
            print(f"TooManyRedirects 에러 발생: {url}")
        except exceptions.RequestException as e:
            print(f"API 요청 중 에러 발생: {e}")

    print("모든 키워드에 대한 이미지 다운로드 완료!")

# 키워드 리스트
keywords = list(keyword_translations.keys())

# 함수 실행
download_images(keywords, client_id, client_secret)
```
