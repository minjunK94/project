```python
def get_path(path):
    x_path = []
    y = []
    for x in os.walk(path):
        if len(x[2])==0:
            continue
        name = os.path.basename(x[0])
        print(name, '사진', len(x[2]), '장')
        for file_name in x[2]:
            image_full_path = path + '/' + name + '/' + file_name
            x_path.append(image_full_path)
            y.append(name)
    return x_path, y
```

```python
def check_images_in_directory(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                with Image.open(file_path) as img:
                    img.verify()  # 이미지 파일 검증
                    print(f"{file}: Valid {img.format}")
            except Exception as e:
                print(f"{file}: Invalid image - {e}")

# Train 디렉토리 확인
train_dir = r"C:\minjun\project\source\Fish_Catch\img\train"
check_images_in_directory(train_dir)

# Validation 디렉토리 확인
val_dir = r"C:\minjun\project\source\Fish_Catch\img\val"
check_images_in_directory(val_dir)
```

```python
def clean_directory(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                with Image.open(file_path) as img:
                    img.verify()  # 파일이 정상적인 이미지인지 확인
            except Exception as e:
                print(f"Removing invalid file: {file_path} ({e})")
                os.remove(file_path)

# 디렉토리 정리
clean_directory(train_dir)
clean_directory(val_dir)
```

```python

def convert_images_to_jpeg(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                with Image.open(file_path) as img:
                    img = img.convert("RGB")  # RGB 모드로 변환
                    new_file_path = os.path.splitext(file_path)[0] + ".jpg"
                    img.save(new_file_path, format="JPEG")
                    if file_path != new_file_path:
                        os.remove(file_path)  # 원본 파일 삭제
            except Exception as e:
                print(f"Skipping invalid file: {file_path} ({e})")

# JPEG 변환
convert_images_to_jpeg(train_dir)
convert_images_to_jpeg(val_dir)
```
